# 歌詞詞頻統計

## Description

你想分析某個樂團的作詞風格，你想先從他使用詞語的頻率(Term Frequency)入手。給你一段歌詞，試著分析單字出現個數。

## Input

輸入共數行，為欲分析的整段歌詞，其中換行以"\n"代替，除了英文字母外，歌詞中亦包含許多標點符號(這回你只需考慮以下4種符號: ",", ":", "?",".")。而當遇到某一行為'==='則表示此行為最終行，代表讀取結束(若還有歌詞則不予理會)，並且此行'==='不列入詞頻統計。

## Output

輸出格式請滿足以下要求：

詞頻計算不包含標點符號(此題只需考慮4種)與換行符號("\n")。
計算好的詞頻請以英文字母做排序輸出。
請將縮寫還原後再計算詞頻(此題只需考慮以下2種: We're -> We are, I'd -> I would)。
請將所有字母轉成小寫。

```python
lyrics = []
while True:
    i = input()
    if i == '===':
        break
    i = i.split()
    for j in i:
        j = j.lower()
        if j == "we're":
            lyrics.append('we')
            lyrics.append('are')
            continue
        if j == "i'd":
            lyrics.append('i')
            lyrics.append('would')
            continue
        j = j.replace('.', '').replace(':', '').replace('?', '').replace(',', '').replace('\n', '')
        lyrics.append(j)
d = {}
for i in sorted(lyrics):
    if i in d:
        d[i] += 1
        continue
    d[i] = 1
for key, value in d.items():
    print('{}:{}'.format(key,value))
```

H1:M1 H2:M2
